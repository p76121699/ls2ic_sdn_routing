```python
#===================== Parameter settings =========================
algs_name = "ls2ic"
num_agents = 992 #32*31, geant topolopy is 506 = 23*22
```

```python
import numpy as np
import matplotlib.pyplot as plt 


def get_smooth(l, s_v):
    s_l = []
    for i in range(len(l)):
        s_l.append(l[i] * (1 - s_v) + s_l[i-1] * s_v if i > 0 else l[i])
    return s_l

plt.rcParams.update({'font.size': 14})

if __name__ == '__main__':
    plt_r = []
    with open(f'../results/{algs_name}/output.txt', 'r') as f1:
        for line in f1:
            token = line.strip()
            if token:                 
                value = float(token)  
                plt_r.append(value / num_agents)

    plt.figure(figsize=(10,10))
    plt.xlabel('Step')
    plt.ylabel('Average reward')
    plt.plot(np.arange(len(plt_r)), plt_r, alpha = 0.2, color='C0')
    plt_r = get_smooth(plt_r,0.95)
    plt.plot(np.arange(len(plt_r)), plt_r, alpha = 0.5, color='blue')  
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 


def get_smooth(l, s_v):
    s_l = []
    for i in range(len(l)):
        s_l.append(l[i] * (1 - s_v) + s_l[i-1] * s_v if i > 0 else l[i])
    return s_l


if __name__ == '__main__':
    plt_r = []
    with open(f'../results/{algs_name}/dqn_loss.txt', 'r') as f1:
        for line in f1:
            token = line.strip() 
            if not token or token.lower() == "none":
                continue
            try:
                value = float(token)
                plt_r.append(value)
            except ValueError:
                continue

    plt.figure(figsize=(10,10))
    plt.xlabel('Step')
    plt.ylabel('Average td loss')
    plt.plot(np.arange(len(plt_r)), plt_r, alpha = 0.2, color='blue')  
    plt_r = get_smooth(plt_r,0.95)
    plt.plot(np.arange(len(plt_r)), plt_r, alpha = 0.5, color='blue')  
    plt.show()
```
